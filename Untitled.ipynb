```python
import json
import emoji
import pandas as pd
```

```python
import datetime
```

```python
PATH_TO_JSON = "data/result.json"
```

```python
EMOJIS = emoji.UNICODE_EMOJI["en"].keys()
```

```python
with open(PATH_TO_JSON, encoding='utf-8') as user_data_json:    
    data = json.load(user_data_json)
    
chats_df = pd.json_normalize(data["chats"]["list"])
```

```python
chats_df.messages = chats_df.messages.apply(pd.DataFrame)
```

```python
pd.options.display.max_columns = 453
```

```python
for i, line in chats_df.iterrows():
    line.messages["chat_name"] = line["name"]
    line.messages["chat_id"] = line.id
    line.messages["pm"] = line.type == "personal_chat"
```

```python
messages_df = pd.concat(chats_df.messages.tolist())
```

```python
messages_df.date = pd.to_datetime(messages_df.date)
```

```python
messages_df.photo = ~messages_df.photo.isna()
```

```python
messages_df.file = ~messages_df.file.isna()
```

```python
important_cols = ['id', 'type', 'date', 'date_unixtime', 'actor', 'actor_id', 'action', 'text', 'from', 'from_id', 'reply_to_message_id', 'photo', 'file', 'media_type', 'sticker_emoji', 'forwarded_from', 'chat_name', 'chat_id', 'pm', 'inviter']
```

```python
messages_df = messages_df[important_cols]
```

```python
def contains_link(message):
    if type(message) == list:
        for i in message:
            if type(i) == dict and i["type"] == "link":
                return True
    return False
```

```python
messages_df["link"] = messages_df.text.apply(contains_link)
```

```python
messages_df.text = messages_df.text.apply(lambda x: " ".join([i if type(i) == str else i["text"] for i in x]) if type(x) == list else x)
```

```python
messages_df["question_mark"] = messages_df.text.str.contains("\?|՞")
```

```python
messages_df["emoji_list"] = messages_df.text.apply(lambda x: {i for i in x if i in EMOJIS})
messages_df["emoji"] = messages_df.emoji_list.apply(len).astype(bool)
```

```python
messages_df.text = messages_df.text.str.replace("[՞՛՜]+", "").str.replace("\W+", " ")
```

```python
def dayify(date, day_start=6):
    return date.date() - datetime.timedelta(days=1) if date.hour <= day_start else date.date()
```

```python
messages_df["day_session"] = messages_df.date.apply(dayify)
```

```python
words_df = messages_df[(messages_df.text != "") & (messages_df.pm)][["id", "chat_id", "from", "day_session", "text"]]
```

```python
words_df.text = words_df.text.str.split()
```
